```python
import pandas as pd
import numpy as np

from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import re
```

```python
train_data = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
test_data = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
```

```python
#QUESTION1:
data = pd.concat([train_data, test_data])
```

```python
#QUESTION1:
data.head(5)
```

```python
#QUESTION2:
data['Survived'] = data['Survived'].dropna()
data['Survived'].value_counts()
```

```python
#QUESTION3:
data['Has_Cabin'] = ~data['Cabin'].isnull()
data.head(5)
```

```python
#QUESTION4:
data['Title'] = data.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
data['Title'] = data['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
data['Title'] = data['Title'].replace(['Don', 'Dona', 'Rev', 'Dr',
                                            'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')
```

```python
#QUESTION5:
data['Age'] = data.Age.fillna(data.Age.median())
data['Fare'] = data.Fare.fillna(data.Fare.median())
data['Embarked'] = data['Embarked'].fillna('S')
```

```python
data.head(5)
```

```python
#QUESTION6:
data['Catagorical_Age'] = pd.qcut(data.Age, q=4, labels=False)
data['Catagorical_Fare'] = pd.qcut(data.Fare, q=4, labels=False)
data.head()
```

```python
#QUESTION7:
data.drop(['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'], axis=1, inplace=True)
data.head()
```

```python
#QUESTION8:
df = pd.get_dummies(data, drop_first=True)
df.head()
```

```python
x = df.drop(['Survived'], axis=1).values
y = df['Survived'].values
```

```python
x.shape
```

```python
y.shape
```

```python
#QUESTION9:
X_train,X_test,Y_train,Y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print("X_train: {}".format(X_train.shape))
print("Y_train: {}".format(Y_train.shape))
print("X_test: {}".format(X_test.shape))
print("Y_test: {}".format(Y_test.shape))
```

```python
#QUESTION10:
le = linear_model.LogisticRegression()
```

```python
le.fit(X_train, Y_train)
```

```python
predict = le.predict(X_test)
```

```python
#Question11:
accuracy_score = le.score(X_test, Y_test)
```

```python
print("Accuracy Score: {}".format(accuracy_score))
```
