```python
import numpy as np
```

```python
#Q1: Mean of Elements
a = np.random.random((10,1))  #Stores the 10 random variables in "a" array
np.mean(a)                    #Calculates the mean of elements
```

```python
#Q2: Variance and Standard Deviation
b = np.random.random((20,1))   #Stores the 20 random variables in "b" array 
sd = np.std(b)                      #Calculate the Standard Deviation 
var = np.var(b)                     #Calculate the Variance
print("Standard Deviation is: ",sd)
print("Variance is: ",var)
```

```python
#Q3: Sum of matix elements
A = np.random.random((10,20))  #Creates array "A"
B = np.random.random((20,25))  #Creates array "B"
C = np.matmul(A,B)             #Matrix multiplication of A and B
print(C)
print(np.shape(C))             
print(C.sum())
```

```python
#Q4: Array of required size
c = np.random.random((10,1))
print(c)

def arrayfunc(x):
    return(1/(1+np.exp(-x)))

result = np.apply_along_axis(arrayfunc, 0, c)
print("Resultant array is: ",result)
```
