```python
#Question3: Univariate Analysis
data_url = "https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv"
house = pd.read_csv(data_url)
bplot = sns.boxplot(y='price', x='bedrooms', data=house, width=0.5, palette="colorblind")
```

```python
#list of outliers
Q1 = house.quantile(0.25)
Q3 = house.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
#QUESTION4:Bivariate Analysis
from matplotlib import pyplot
from numpy import mean
from scipy.stats import spearmanr
x = "price"
y = "bedrooms"
pyplot.scatter(x,y)
pyplot.show()
corr, _ = spearmanr(x,y)
print(corr)
```
